# Accuracy in Machine Learning

## What is Accuracy?

Accuracy is one of the most commonly used classification evaluation metrics.

It tells us:

> **Out of all predictions made by the model, how many were correct?**

Accuracy measures the overall correctness of a model.

---

# Accuracy Formula

$$
Accuracy = \frac{Correct\ Predictions}{Total\ Predictions}
$$

In terms of a confusion matrix:

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

Where:

- **TP (True Positive)** = Correctly predicted Positive
- **TN (True Negative)** = Correctly predicted Negative
- **FP (False Positive)** = Incorrectly predicted Positive
- **FN (False Negative)** = Incorrectly predicted Negative

---

# Understanding the Formula

The numerator:

$$
TP + TN
$$

represents all correct predictions.

The denominator:

$$
TP + TN + FP + FN
$$

represents all predictions.

Therefore:

$$
Accuracy = \frac{Correct\ Predictions}{Total\ Predictions}
$$

---

# Confusion Matrix

| Actual / Predicted | Positive | Negative |
|----------|----------|----------|
| Positive | TP | FN |
| Negative | FP | TN |

Accuracy uses all four values:

$$
TP
$$

$$
TN
$$

$$
FP
$$

$$
FN
$$

Unlike Precision and Recall, Accuracy considers the entire confusion matrix.

---

# Simple Example

Suppose a model predicts whether an email is Spam or Not Spam.

After testing:

| Actual / Predicted | Spam | Not Spam |
|----------|----------|----------|
| Spam | 40 | 10 |
| Not Spam | 5 | 45 |

So:

$$
TP = 40
$$

$$
FN = 10
$$

$$
FP = 5
$$

$$
TN = 45
$$

---

# Step 1: Apply the Formula

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

Substitute values:

$$
Accuracy = \frac{40 + 45}{40 + 45 + 5 + 10}
$$

$$
Accuracy = \frac{85}{100}
$$

$$
Accuracy = 0.85
$$

$$
Accuracy = 85\%
$$

---

# What Does 85% Accuracy Mean?

An accuracy of 85% means:

> Out of every 100 predictions, the model correctly predicts 85 of them.

The remaining 15 predictions are incorrect.

---

# Another Example

$$

$$

$$

$$

$$

$$

$$

$$

Applying the formula:

---

# Why Accuracy is Popular

Accuracy is:

- Easy to understand
- Easy to calculate
- Gives a quick overview of model performance

Because of this, it is usually the first metric people check.

---

# The Problem with Accuracy

Accuracy can be misleading when classes are imbalanced.

---

# Example of Misleading Accuracy

Imagine a disease dataset with:

- 1000 patients
- 990 healthy patients
- 10 sick patients

Now suppose a model predicts:

> Everyone is Healthy

Then:

$$
TP = 0
$$

$$
FN = 10
$$

$$
FP = 0
$$

$$
TN = 990
$$

Accuracy becomes:

$$
Accuracy = \frac{0 + 990}{1000}
$$

$$
Accuracy = 99\%
$$

This seems excellent.

However:

- The model detected zero sick patients.
- The model is actually useless.

This is why Accuracy alone should not always be trusted.

---

# Accuracy vs Precision

## Accuracy

Question:

> How many total predictions were correct?

Formula:

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

---

## Precision

Question:

> When the model predicts Positive, how often is it correct?

Formula:

$$
Precision = \frac{TP}{TP+FP}
$$

---

# Accuracy vs Recall

## Accuracy

Question:

> How many total predictions were correct?

Formula:

$$
Accuracy = \frac{TP + TN}{TP + TN + FP + FN}
$$

---

## Recall

Question:

> Out of all actual Positives, how many were found?

Formula:

$$
Recall = \frac{TP}{TP+FN}
$$

---

# Accuracy vs F1-Score

## Accuracy

Measures:

$$
Overall\ Correctness
$$

---

## F1-Score

Measures:

$$
Balance\ Between\ Precision\ and\ Recall
$$

Formula:

$$
F1 = 2\times\frac{Precision\times Recall}{Precision+Recall}
$$

---

# When Should We Use Accuracy?

Accuracy works well when:

- Classes are balanced.
- Positive and Negative classes have similar importance.

Examples:

- Digit Recognition
- Object Classification
- Basic Image Classification

---

# When Should We Avoid Accuracy?

Accuracy should not be used alone when:

- Classes are highly imbalanced.
- Missing positives is expensive.

Examples:

- Fraud Detection
- Cancer Detection
- Security Threat Detection
- Rare Disease Detection

In these cases:

- Precision
- Recall
- F1-Score

are usually more important.

# Easy Real-Life Example

Then:

$$
Accuracy = \frac{90}{100}
$$

$$

This means the teacher was correct 90% of the time.

# Key Points

- Accuracy measures overall correctness.
- Accuracy uses all four values of the confusion matrix.
- Easy to calculate and understand.
- Can be misleading for imbalanced datasets.
- Should often be used together with Precision, Recall, and F1-Score.

$$
