# F1-Score in Machine Learning

## What is F1-Score?

F1-Score is a classification evaluation metric that combines:

- Precision
- Recall

into a single score.

F1-Score helps us measure the overall performance of a classification model when both Precision and Recall are important.

---

# Why Do We Need F1-Score?

Suppose a model has:

- Very high Precision
- Very low Recall

or

- Very high Recall
- Very low Precision

Looking at only one metric can be misleading.

F1-Score balances both metrics and gives a single value that reflects the model's overall quality.

---

# F1-Score Formula

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

This formula is known as the **Harmonic Mean** of Precision and Recall.

---

# Understanding the Formula

The formula is:

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

Notice that both Precision and Recall are included.

Therefore:

- High Precision + High Recall = High F1-Score
- Low Precision or Low Recall = Low F1-Score

The model must perform well on both metrics to achieve a high F1-Score.

---

# Why Harmonic Mean?

You might wonder:

Why not simply take the average?

For example:

$$
\frac{Precision + Recall}{2}
$$

The reason is that the Harmonic Mean heavily penalizes low values.

If either Precision or Recall is very low, the F1-Score will also become low.

This prevents a model from appearing good when one metric is poor.

---

# Simple Example

Suppose:

$$
Precision = 0.80
$$

$$
Recall = 0.60
$$

Applying the formula:

$$
F1 = 2 \times \frac{0.80 \times 0.60}{0.80 + 0.60}
$$

$$
F1 = 2 \times \frac{0.48}{1.40}
$$

$$
F1 = \frac{0.96}{1.40}
$$

$$
F1 = 0.686
$$

$$
F1 = 68.6\%
$$

---

# Another Example

Suppose:

$$
Precision = 90\%
$$

$$
Recall = 90\%
$$

Then:

$$
F1 = 2 \times \frac{0.90 \times 0.90}{0.90 + 0.90}
$$

$$
F1 = 0.90
$$

$$
F1 = 90\%
$$

Since both metrics are high, the F1-Score is also high.

---

# Example with Low Recall

Suppose:

$$
Precision = 95\%
$$

$$
Recall = 20\%
$$

Applying the formula:

$$
F1 = 2 \times \frac{0.95 \times 0.20}{0.95 + 0.20}
$$

$$
F1 = 0.33
$$

$$
F1 = 33\%
$$

Even though Precision is excellent, the F1-Score is poor because Recall is very low.

This shows that F1-Score punishes imbalance.

---

# Confusion Matrix Example

| Actual / Predicted | Positive | Negative |
|----------|----------|----------|
| Positive | TP = 40 | FN = 10 |
| Negative | FP = 20 | TN = 30 |

---

## Step 1: Calculate Precision

$$
Precision = \frac{TP}{TP + FP}
$$

$$
Precision = \frac{40}{40+20}
$$

$$
Precision = \frac{40}{60}
$$

$$
Precision = 0.667
$$

---

## Step 2: Calculate Recall

$$
Recall = \frac{TP}{TP + FN}
$$

$$
Recall = \frac{40}{40+10}
$$

$$
Recall = \frac{40}{50}
$$

$$
Recall = 0.80
$$

---

## Step 3: Calculate F1-Score

$$
F1 = 2 \times \frac{0.667 \times 0.80}{0.667 + 0.80}
$$

$$
F1 = 0.727
$$

$$
F1 = 72.7\%
$$

---

# Range of F1-Score

The F1-Score always lies between:

$$
0 \leq F1 \leq 1
$$

or

$$
0\% \leq F1 \leq 100\%
$$

---

## Worst Possible Score

$$
F1 = 0
$$

The model performs very poorly.

---

## Best Possible Score

$$
F1 = 1
$$

or

$$
100\%
$$

The model has perfect Precision and Recall.

---

# When Should We Use F1-Score?

F1-Score is useful when:

- Classes are imbalanced.
- Precision and Recall are equally important.
- We need a single performance metric.

Examples:

- Fraud Detection
- Disease Detection
- Spam Detection
- Cyber Security
- Fake News Detection

---

# Precision vs Recall vs F1-Score

## Precision

Question:

> When the model predicts Positive, how often is it correct?

Formula:

$$
Precision = \frac{TP}{TP+FP}
$$

Focuses on:

$$
False\ Positives
$$

---

## Recall

Question:

> Out of all actual Positive cases, how many did the model find?

Formula:

$$
Recall = \frac{TP}{TP+FN}
$$

Focuses on:

$$
False\ Negatives
$$

---

## F1-Score

Question:

> How good is the balance between Precision and Recall?

Formula:

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

Focuses on:

- Precision
- Recall

together.

---

# Easy Real-Life Example

Imagine a disease detection system.

The model:

- Correctly detects many sick people.
- Makes very few wrong disease predictions.

This means:

$$
High\ Precision
$$

and

$$
High\ Recall
$$

Therefore:

$$
High\ F1\ Score
$$

Now imagine:

- It detects only a few sick patients.

Then Recall becomes low.

Even if Precision remains high, F1-Score drops significantly.

---

# Easy Way to Remember F1-Score

Think:

> "F1-Score is the balance score between Precision and Recall."

Precision asks:

> "How accurate are my positive predictions?"

Recall asks:

> "How many real positive cases did I find?"

F1-Score asks:

> "How good are both of them together?"

---

# Key Points

- F1-Score combines Precision and Recall.
- It uses the Harmonic Mean.
- High F1-Score requires both Precision and Recall to be high.
- If either Precision or Recall is low, F1-Score decreases.
- Useful for imbalanced datasets.
- Commonly used in classification problems.

Formula:

$$
F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

Remember:

> F1-Score is the balance between Precision and Recall.
