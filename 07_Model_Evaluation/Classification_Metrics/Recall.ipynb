# Recall in Machine Learning

## What is Recall?

Recall is a classification evaluation metric that tells us:

> **Out of all actual positive cases, how many did the model correctly identify?**

Recall focuses on finding all positive cases.

It is especially important when:

- Missing a positive case is very costly.
- We want to catch as many positive cases as possible.

Examples:

- Disease detection
- Fraud detection
- Spam detection
- Security threat detection

---

# Recall Formula

$$
Recall = \frac{TP}{TP + FN}
$$

Where:

- **TP (True Positive)** = Model correctly predicts Positive.
- **FN (False Negative)** = Model predicts Negative but actual is Positive.

---

# Understanding the Formula

$$
Recall = \frac{\text{Correctly Predicted Positives}}{\text{All Actual Positives}}
$$

or

$$
Recall = \frac{TP}{TP + FN}
$$

Notice:

$$
TP + FN
$$

represents all actual positive cases.

So Recall measures:

> "How many actual positive cases were successfully found by the model?"

---

# Simple Example

Suppose a hospital has:

- 100 patients
- 20 patients actually have a disease

After prediction:

- Model correctly detects 15 sick patients
- Model misses 5 sick patients

So:

$$
TP = 15
$$

$$
FN = 5
$$

Applying the formula:

$$
Recall = \frac{15}{15 + 5}
$$

$$
Recall = \frac{15}{20}
$$

$$
Recall = 0.75
$$

$$
Recall = 75\%
$$

---

# What Does 75% Recall Mean?

A recall of 75% means:

> Out of all people who actually had the disease, the model successfully found 75% of them.

The remaining 25% were missed by the model.

---

# Confusion Matrix Example

| Actual / Predicted | Positive | Negative |
|----------|----------|----------|
| Positive | TP = 15 | FN = 5 |
| Negative | FP = 10 | TN = 70 |

From the confusion matrix:

$$
Recall = \frac{TP}{TP + FN}
$$

$$
Recall = \frac{15}{15 + 5}
$$

$$
Recall = 75\%
$$

Notice that:

- FP does not affect Recall.
- TN does not affect Recall.

Recall only cares about:

$$
TP
$$

and

$$
FN
$$

---

# Why Recall is Important

Imagine a cancer detection system.

There are 100 cancer patients.

The model finds:

- 95 patients correctly
- Misses 5 patients

Then:

$$
Recall = \frac{95}{95+5}
$$

$$
Recall = 95\%
$$

This is good because very few patients were missed.

Now suppose another model finds only:

- 50 patients correctly
- Misses 50 patients

Then:

$$
Recall = \frac{50}{50+50}
$$

$$
Recall = 50\%
$$

This is dangerous because half of the patients were not detected.

---

# High Recall vs Low Recall

## High Recall

- Finds most positive cases.
- Very few positive cases are missed.
- Lower False Negatives.

Example:

$$
Recall = 95\%
$$

Only a few positive cases are missed.

---

## Low Recall

- Misses many positive cases.
- High False Negatives.

Example:

$$
Recall = 40\%
$$

Many positive cases remain undetected.

---

# Recall and False Negatives

Recall is strongly related to False Negatives.

The formula is:

$$
Recall = \frac{TP}{TP + FN}
$$

If FN increases:

$$
Recall \downarrow
$$

If FN decreases:

$$
Recall \uparrow
$$

Therefore:

> Lower False Negatives = Higher Recall
