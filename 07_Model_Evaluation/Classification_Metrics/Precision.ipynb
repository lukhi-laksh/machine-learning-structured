# Precision in Machine Learning

## What is Precision?

Precision is a classification evaluation metric that tells us:

> **Out of all cases predicted as Positive, how many were actually Positive?**

Precision focuses on the quality of positive predictions.

It answers the question:

> "When the model says YES, how often is it correct?"

---

# Precision Formula

$$
Precision = \frac{TP}{TP + FP}
$$

Where:

- **TP (True Positive)** = Model correctly predicts Positive.
- **FP (False Positive)** = Model predicts Positive but actual is Negative.

---

# Understanding the Formula

$$
Precision = \frac{\text{Correct Positive Predictions}}{\text{All Positive Predictions}}
$$

or

$$
Precision = \frac{TP}{TP + FP}
$$

Notice:

$$
TP + FP
$$

represents all samples that the model predicted as Positive.

Therefore Precision measures:

> "Out of everything the model labeled Positive, how much was actually Positive?"

---

# Simple Example

Suppose a model is used to detect fraud.

The model predicts 20 transactions as Fraud.

After verification:

- 15 were actually Fraud.
- 5 were actually Normal transactions.

So:

$$
TP = 15
$$

$$
FP = 5
$$

Applying the formula:

$$
Precision = \frac{15}{15+5}
$$

$$
Precision = \frac{15}{20}
$$

$$
Precision = 0.75
$$

$$
Precision = 75\%
$$

---

# What Does 75% Precision Mean?

A precision of 75% means:

> Out of all transactions predicted as Fraud, 75% were truly Fraud.

The remaining 25% were wrong predictions.

---

# Confusion Matrix Example

| Actual / Predicted | Positive | Negative |
|----------|----------|----------|
| Positive | TP = 15 | FN = 5 |
| Negative | FP = 5 | TN = 75 |

Using the formula:

$$
Precision = \frac{TP}{TP + FP}
$$

$$
Precision = \frac{15}{15+5}
$$

$$
Precision = 75\%
$$

Notice that:

- FN does not affect Precision.
- TN does not affect Precision.

Precision only cares about:

$$
TP
$$

and

$$
FP
$$

---

# Why Precision is Important

Imagine an email spam filter.

Suppose:

- The model marks 100 emails as Spam.
- Only 20 are truly Spam.
- The other 80 are important emails.

Then:

$$
Precision = \frac{20}{20+80}
$$

$$
Precision = 20\%
$$

This is bad because most spam predictions were wrong.

Now consider:

- Model marks 100 emails as Spam.
- 95 are truly Spam.
- Only 5 are normal emails.

Then:

$$
Precision = \frac{95}{95+5}
$$

$$
Precision = 95\%
$$

This is excellent because almost every spam prediction was correct.

---

# High Precision vs Low Precision

## High Precision

- Very few False Positives.
- Positive predictions are usually correct.
- Model is trustworthy when it predicts Positive.

Example:

$$
Precision = 95\%
$$

Only a few wrong positive predictions.

---

## Low Precision

- Many False Positives.
- Positive predictions are often wrong.

Example:

$$
Precision = 40\%
$$

Many predicted positives are actually negative.

---

# Precision and False Positives

Precision is strongly related to False Positives.

The formula is:

$$
Precision = \frac{TP}{TP + FP}
$$

If FP increases:

$$
Precision \downarrow
$$

If FP decreases:

$$
Precision \uparrow
$$

Therefore:

> Lower False Positives = Higher Precision

# Precision vs Recall

Precision asks:

> "When the model predicts Positive, how often is it correct?"

Formula:

$$
Precision = \frac{TP}{TP + FP}
$$

Recall asks:

> "Out of all actual Positive cases, how many did the model find?"

Formula:

$$
Recall = \frac{TP}{TP + FN}
$$

---

# Easy Example to Understand Precision and Recall

Suppose there are:

- 100 patients
- 20 actually have a disease

The model predicts 30 patients as Sick.

Among those:

- 15 are actually Sick
- 15 are actually Healthy

So:

$$
TP = 15
$$

$$
FP = 15
$$

$$
FN = 5
$$

### Precision

$$
Precision = \frac{15}{15+15}
$$

$$
Precision = 50\%
$$

Meaning:

> When the model says someone is Sick, it is correct only 50% of the time.

### Recall

$$
Recall = \frac{15}{15+5}
$$

$$
Recall = 75\%
$$

Meaning:

> The model found 75% of all sick patients.

# Easy Way to Remember Precision

> "Precision measures how precise or accurate the model's positive predictions are."

Formula:

$$
Precision = \frac{TP}{TP + FP}
$$

Focuses on:

- Predicted Positives
- Wrong Positive Predictions (FP)

Main Question:

> "Out of everything predicted Positive, how many were actually Positive?"

# Key Points

- Precision is a classification metric.
- Precision measures the quality of positive predictions.
- High Precision means fewer False Positives.
- Precision focuses on TP and FP only.
- Precision is important when wrong positive predictions are costly.

Formula:

$$
Precision = \frac{TP}{TP + FP}
$$

Remember:

> High Precision = Fewer False Positives.
