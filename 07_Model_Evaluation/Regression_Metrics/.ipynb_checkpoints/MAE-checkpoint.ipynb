# Mean Absolute Error (MAE)

## Definition

Mean Absolute Error (MAE) is a regression evaluation metric that measures the average absolute difference between actual values and predicted values.

It tells us how far, on average, the predicted values are from the actual values.

---

## Formula

$$
MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|
$$

Where:

- $y_i$ = Actual value
- $\hat{y}_i$ = Predicted value
- $n$ = Number of observations
- $|y_i - \hat{y}_i|$ = Absolute Error

---

## Example

Suppose a model predicts house prices.

| Actual Price | Predicted Price |
|-------------|----------------|
| 100 | 90 |
| 150 | 170 |
| 200 | 180 |
| 250 | 260 |

### Step 1: Calculate Errors

| Actual | Predicted | Error |
|----------|----------|----------|
| 100 | 90 | 10 |
| 150 | 170 | -20 |
| 200 | 180 | 20 |
| 250 | 260 | -10 |

### Step 2: Calculate Absolute Errors

| Error | Absolute Error |
|---------|---------|
| 10 | 10 |
| -20 | 20 |
| 20 | 20 |
| -10 | 10 |

### Step 3: Calculate MAE

$$
MAE = \frac{10 + 20 + 20 + 10}{4}
$$

$$
MAE = \frac{60}{4}
$$

$$
MAE = 15
$$

---

## Interpretation

The MAE value is **15**, which means the model's predictions are off by an average of **15 units** from the actual values.

---

## Python Example

```python
from sklearn.metrics import mean_absolute_error

actual = [100, 150, 200, 250]
predicted = [90, 170, 180, 260]

mae = mean_absolute_error(actual, predicted)

print("MAE:", mae)
```

Output:

```python
MAE: 15.0
```

---

## Advantages of MAE

- Easy to understand and interpret.
- Uses the same unit as the target variable.
- Less sensitive to outliers than RMSE.
- Suitable for business reporting and forecasting.

---

## Disadvantages of MAE

- Does not heavily penalize large errors.
- Different models can have the same MAE but different error distributions.

---

## Real-World Example (Loom Efficiency Forecasting)

| Actual Efficiency | Predicted Efficiency |
|------------------|---------------------|
| 85 | 83 |
| 90 | 88 |
| 75 | 80 |

Absolute Errors:

$$
|85-83| = 2
$$

$$
|90-88| = 2
$$

$$
|75-80| = 5
$$

$$
MAE = \frac{2+2+5}{3}
$$

$$
MAE = 3
$$

This means the forecasting model misses the actual efficiency by approximately **3 efficiency points on average**.
