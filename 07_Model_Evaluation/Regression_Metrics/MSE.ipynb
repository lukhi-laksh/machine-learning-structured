# Mean Squared Error (MSE)

## Definition

Mean Squared Error (MSE) is a regression evaluation metric that measures the average of the squared differences between actual values and predicted values.

MSE penalizes larger errors more heavily because the errors are squared before averaging.

---

## Formula

$$
MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

Where:

- $y_i$ = Actual value
- $\hat{y}_i$ = Predicted value
- $n$ = Number of observations
- $(y_i - \hat{y}_i)^2$ = Squared Error

---

## Example

Suppose a model predicts house prices.

| Actual Price | Predicted Price |
|-------------|----------------|
| 100 | 90 |
| 150 | 170 |
| 200 | 180 |
| 250 | 260 |

### Step 1: Calculate Errors

| Actual | Predicted | Error |
|----------|----------|----------|
| 100 | 90 | 10 |
| 150 | 170 | -20 |
| 200 | 180 | 20 |
| 250 | 260 | -10 |

### Step 2: Square Each Error

| Error | Squared Error |
|---------|---------|
| 10 | 100 |
| -20 | 400 |
| 20 | 400 |
| -10 | 100 |

### Step 3: Calculate MSE

$$
MSE = \frac{100 + 400 + 400 + 100}{4}
$$

$$
MSE = \frac{1000}{4}
$$

$$
MSE = 250
$$

---

## Interpretation

The MSE value is **250**, which means the average squared prediction error is 250.

A lower MSE indicates better model performance.

---

## Python Example

```python
from sklearn.metrics import mean_squared_error

actual = [100, 150, 200, 250]
predicted = [90, 170, 180, 260]

mse = mean_squared_error(actual, predicted)

print("MSE:", mse)
```

Output:

```python
MSE: 250.0
```

---

## Advantages of MSE

- Easy to calculate.
- Widely used in machine learning.
- Penalizes large errors heavily.
- Useful for optimization algorithms.

---

## Disadvantages of MSE

- Sensitive to outliers.
- Unit becomes squared (e.g., meters², dollars²).
- Harder to interpret than MAE.

---

## Difference Between MAE and MSE

| Metric | Formula | Large Errors |
|----------|----------|----------|
| MAE | Average Absolute Error | Penalizes equally |
| MSE | Average Squared Error | Penalizes heavily |

---

## Real-World Example (Loom Efficiency Forecasting)

| Actual Efficiency | Predicted Efficiency |
|------------------|---------------------|
| 85 | 83 |
| 90 | 88 |
| 75 | 80 |

### Calculate Squared Errors

$$
(85 - 83)^2 = 4
$$

$$
(90 - 88)^2 = 4
$$

$$
(75 - 80)^2 = 25
$$

### Calculate MSE

$$
MSE = \frac{4 + 4 + 25}{3}
$$

$$
MSE = \frac{33}{3}
$$

$$
MSE = 11
$$

This means the average squared prediction error of the forecasting model is **11**.

---

## Key Point

- **MAE** tells the average error directly.
- **MSE** tells the average squared error and penalizes large mistakes much more.
- Machine learning algorithms such as Linear Regression often use MSE as their default loss function.
