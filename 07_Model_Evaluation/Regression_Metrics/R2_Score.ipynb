# R² Score (Coefficient of Determination)

## Definition

R² Score (R-Squared) is a regression evaluation metric that measures how well the model explains the variability in the target variable.

It indicates the proportion of variance in the dependent variable that can be predicted from the independent variables.

---

## Formula

$$
R^2 = 1 - \frac{\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}{\sum_{i=1}^{n}(y_i - \bar{y})^2}
$$

Where:

- $y_i$ = Actual value
- $\hat{y}_i$ = Predicted value
- $\bar{y}$ = Mean of actual values
- $n$ = Number of observations

---

## Understanding the Formula

The numerator represents the **Residual Sum of Squares (RSS)**:

$$
RSS = \sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

The denominator represents the **Total Sum of Squares (TSS)**:

$$
TSS = \sum_{i=1}^{n}(y_i - \bar{y})^2
$$

Therefore,

$$
R^2 = 1 - \frac{RSS}{TSS}
$$

---

## Example

Suppose we have:

| Actual | Predicted |
|----------|----------|
| 10 | 12 |
| 20 | 18 |
| 30 | 29 |
| 40 | 41 |

### Step 1: Calculate Mean of Actual Values

$$
\bar{y} = \frac{10 + 20 + 30 + 40}{4}
$$

$$
\bar{y} = 25
$$

---

### Step 2: Calculate RSS

$$
RSS = (10-12)^2 + (20-18)^2 + (30-29)^2 + (40-41)^2
$$

$$
RSS = 4 + 4 + 1 + 1
$$

$$
RSS = 10
$$

---

### Step 3: Calculate TSS

$$
TSS = (10-25)^2 + (20-25)^2 + (30-25)^2 + (40-25)^2
$$

$$
TSS = 225 + 25 + 25 + 225
$$

$$
TSS = 500
$$

---

### Step 4: Calculate R² Score

$$
R^2 = 1 - \frac{10}{500}
$$

$$
R^2 = 1 - 0.02
$$

$$
R^2 = 0.98
$$

---

## Final Answer

$$
R^2 = 0.98
$$

or

$$
R^2 = 98\%
$$

This means the model explains **98% of the variance** in the target variable.

---

## Interpretation

| R² Score | Meaning |
|----------|----------|
| 1.0 | Perfect Prediction |
| 0.9 - 1.0 | Excellent Model |
| 0.7 - 0.9 | Good Model |
| 0.5 - 0.7 | Moderate Model |
| 0.0 | No Improvement Over Mean |
| Less than 0 | Worse than Predicting the Mean |

---

## Python Example

```python
from sklearn.metrics import r2_score

actual = [10, 20, 30, 40]
predicted = [12, 18, 29, 41]

r2 = r2_score(actual, predicted)

print("R² Score:", r2)
```

Output:

```python
R² Score: 0.98
```

---

## Advantages of R² Score

- Easy to interpret.
- Measures how well the model fits the data.
- Useful for comparing regression models.
- Scale-independent metric.

---

## Disadvantages of R² Score

- Does not indicate whether predictions are biased.
- Can be misleading when used alone.
- Adding more features can artificially increase R².

---

## Difference Between MAE, MSE and R²

| Metric | Best Value | Meaning |
|----------|----------|----------|
| MAE | 0 | Average Absolute Error |
| MSE | 0 | Average Squared Error |
| R² | 1 | Variance Explained by Model |

---

## Real-World Example (Loom Efficiency Forecasting)

Suppose your model predicts loom efficiency.

- Actual efficiencies: 85, 90, 75, 80
- Predicted efficiencies: 84, 89, 77, 79

After calculation:

$$
R^2 = 0.95
$$

This means the model explains **95% of the variation in loom efficiency**, indicating an excellent forecasting model.

---

## Key Point

- **MAE** tells how much the model is wrong on average.
- **MSE** tells how much the squared error is on average.
- **R² Score** tells how well the model explains the target variable.

A good regression model generally has:

$$
MAE \rightarrow 0
$$

$$
MSE \rightarrow 0
$$

$$
R^2 \rightarrow 1
$$
