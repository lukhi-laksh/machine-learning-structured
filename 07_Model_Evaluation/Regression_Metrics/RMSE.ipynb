# Root Mean Squared Error (RMSE)

## Definition

Root Mean Squared Error (RMSE) is a regression evaluation metric that measures the average magnitude of prediction errors.

RMSE is calculated by taking the square root of the Mean Squared Error (MSE).

It indicates how far, on average, the predicted values are from the actual values.

---

## Formula

$$
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}
$$

Where:

- $y_i$ = Actual value
- $\hat{y}_i$ = Predicted value
- $n$ = Number of observations

---

## Relationship Between MSE and RMSE

First calculate MSE:

$$
MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2
$$

Then take the square root:

$$
RMSE = \sqrt{MSE}
$$

---

## Example

Suppose a model predicts house prices.

| Actual Price | Predicted Price |
|-------------|----------------|
| 100 | 90 |
| 150 | 170 |
| 200 | 180 |
| 250 | 260 |

### Step 1: Calculate Errors

| Actual | Predicted | Error |
|----------|----------|----------|
| 100 | 90 | 10 |
| 150 | 170 | -20 |
| 200 | 180 | 20 |
| 250 | 260 | -10 |

---

### Step 2: Square the Errors

| Error | Squared Error |
|---------|---------|
| 10 | 100 |
| -20 | 400 |
| 20 | 400 |
| -10 | 100 |

---

### Step 3: Calculate MSE

$$
MSE = \frac{100 + 400 + 400 + 100}{4}
$$

$$
MSE = \frac{1000}{4}
$$

$$
MSE = 250
$$

---

### Step 4: Calculate RMSE

$$
RMSE = \sqrt{250}
$$

$$
RMSE \approx 15.81
$$

---

## Final Answer

$$
RMSE \approx 15.81
$$

This means the model's predictions are off by approximately **15.81 units on average**.

---

## Interpretation

| RMSE Value | Meaning |
|------------|----------|
| 0 | Perfect Prediction |
| Small RMSE | Better Model |
| Large RMSE | Poor Model |

The lower the RMSE, the better the model performance.

---

## Python Example

```python
from sklearn.metrics import root_mean_squared_error

actual = [100, 150, 200, 250]
predicted = [90, 170, 180, 260]

rmse = root_mean_squared_error(actual, predicted)

print("RMSE:", rmse)
```

Output:

```python
RMSE: 15.811388300841896
```

---

## Alternative Method

```python
from sklearn.metrics import mean_squared_error
import numpy as np

actual = [100, 150, 200, 250]
predicted = [90, 170, 180, 260]

rmse = np.sqrt(mean_squared_error(actual, predicted))

print("RMSE:", rmse)
```

---

## Advantages of RMSE

- Same unit as the target variable.
- Easy to interpret.
- Penalizes large errors more than MAE.
- Widely used in forecasting and regression tasks.

---

## Disadvantages of RMSE

- Sensitive to outliers.
- A few large errors can significantly increase RMSE.
- Can sometimes exaggerate model error.

---

## Difference Between MAE, MSE and RMSE

| Metric | Best Value | Unit |
|----------|----------|----------|
| MAE | 0 | Same as target |
| MSE | 0 | Squared unit |
| RMSE | 0 | Same as target |

---

## Real-World Example (Loom Efficiency Forecasting)

Suppose:

| Actual Efficiency | Predicted Efficiency |
|------------------|---------------------|
| 85 | 83 |
| 90 | 88 |
| 75 | 80 |

### Calculate Squared Errors

$$
(85-83)^2 = 4
$$

$$
(90-88)^2 = 4
$$

$$
(75-80)^2 = 25
$$

### Calculate MSE

$$
MSE = \frac{4+4+25}{3}
$$

$$
MSE = 11
$$

### Calculate RMSE

$$
RMSE = \sqrt{11}
$$

$$
RMSE \approx 3.32
$$

This means the model's efficiency predictions are off by approximately **3.32 efficiency points on average**.

---

## Key Point

- **MAE** measures average absolute error.
- **MSE** measures average squared error.
- **RMSE** measures the square root of average squared error.
- **R² Score** measures how much variance is explained by the model.

A good regression model generally has:

$$
MAE \rightarrow 0
$$

$$
MSE \rightarrow 0
$$

$$
RMSE \rightarrow 0
$$

$$
R^2 \rightarrow 1
$$
