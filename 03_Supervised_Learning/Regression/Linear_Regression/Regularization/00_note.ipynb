# Lasso Regression and Ridge Regression

## Introduction

Lasso Regression and Ridge Regression are advanced versions of Linear Regression that are used to reduce overfitting and improve the performance of machine learning models.

Both techniques belong to a family of methods known as **Regularization Techniques**.

Regularization helps prevent a model from becoming too complex by adding a penalty to the model during training.

The main goal is to create a model that performs well not only on training data but also on unseen data.

---

# Why Do We Need Regularization?

Linear Regression tries to find the best relationship between input features and the target variable.

A Linear Regression model is represented as:

$$
y = b_0 + b_1x_1 + b_2x_2 + \cdots + b_nx_n
$$

where:

- $y$ = Predicted value
- $b_0$ = Intercept
- $b_1,b_2,\dots,b_n$ = Coefficients
- $x_1,x_2,\dots,x_n$ = Features

When a dataset contains many features, Linear Regression may start learning noise and unnecessary patterns from the training data.

This problem is known as **Overfitting**.

An overfitted model performs very well on training data but poorly on new data.

To solve this problem, regularization techniques such as Lasso and Ridge Regression are used.

---

# What is Ridge Regression?

Ridge Regression is a regularized version of Linear Regression that adds a penalty for large coefficient values.

The Ridge Regression cost function is:

$$
RSS + \lambda \sum b_i^2
$$

where:

- $RSS$ = Residual Sum of Squares
- $\lambda$ = Regularization parameter
- $b_i$ = Model coefficients

The purpose of Ridge Regression is to reduce the magnitude of coefficients.

Instead of allowing coefficients to become very large, Ridge Regression pushes them toward zero.

However, Ridge Regression does not make coefficients exactly zero.

---

## How Ridge Regression Works

Suppose a Linear Regression model produces:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 15 |
| Attendance | 12 |
| Sleep Hours | 8 |

After applying Ridge Regression:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 10 |
| Attendance | 8 |
| Sleep Hours | 5 |

The coefficients become smaller, which reduces model complexity and helps prevent overfitting.

---

## Advantages of Ridge Regression

- Reduces overfitting.
- Improves model generalization.
- Works well when many features are useful.
- Handles multicollinearity effectively.
- Keeps all features in the model.

---

## Disadvantages of Ridge Regression

- Does not perform feature selection.
- All features remain in the model.
- Less useful when many features are irrelevant.

---

## When to Use Ridge Regression

Use Ridge Regression when:

- Most features are important.
- Features are highly correlated.
- You want to keep all features.
- The dataset contains multicollinearity.

---

# What is Lasso Regression?

Lasso Regression is another regularized version of Linear Regression.

Lasso stands for:

**Least Absolute Shrinkage and Selection Operator**

The Lasso Regression cost function is:

$$
RSS + \lambda \sum |b_i|
$$

where:

- $RSS$ = Residual Sum of Squares
- $\lambda$ = Regularization parameter
- $b_i$ = Model coefficients

Lasso Regression not only reduces coefficient values but can also make some coefficients exactly zero.

This means that Lasso can automatically remove unnecessary features from the model.

---

## How Lasso Regression Works

Suppose a Linear Regression model produces:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 15 |
| Attendance | 12 |
| Sleep Hours | 8 |
| Shoe Size | 2 |
| Favorite Color | 1 |

After applying Lasso Regression:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 10 |
| Attendance | 8 |
| Sleep Hours | 5 |
| Shoe Size | 0 |
| Favorite Color | 0 |

The last two features are removed from the model because they do not contribute significantly to the prediction.

This process is known as **Feature Selection**.

---

## Advantages of Lasso Regression

- Reduces overfitting.
- Automatically selects important features.
- Removes unnecessary features.
- Creates simpler and more interpretable models.
- Useful for datasets with many features.

---

## Disadvantages of Lasso Regression

- May remove features that contain useful information.
- Can become unstable when features are highly correlated.
- Performance may decrease if too many features are removed.

---

## When to Use Lasso Regression

Use Lasso Regression when:

- The dataset contains many features.
- Some features may be irrelevant.
- Feature selection is required.
- A simpler model is preferred.

---

# Understanding Lambda ($$\lambda$$)

The regularization parameter is represented by:

$$
\lambda
$$

Lambda controls the strength of the penalty.

### Small Lambda

$$
\lambda \approx 0
$$

- Very little regularization.
- Model behaves like Linear Regression.

### Large Lambda

$$
\lambda \gg 0
$$

- Strong regularization.
- Coefficients become smaller.
- Risk of underfitting if Lambda is too large.

---

# Difference Between Linear Regression, Ridge Regression and Lasso Regression

| Property | Linear Regression | Ridge Regression | Lasso Regression |
|----------|----------|----------|----------|
| Reduces Overfitting | ❌ | ✅ | ✅ |
| Shrinks Coefficients | ❌ | ✅ | ✅ |
| Removes Features | ❌ | ❌ | ✅ |
| Feature Selection | ❌ | ❌ | ✅ |
| Handles Multicollinearity | ❌ | ✅ | ✅ |
| Keeps All Features | ✅ | ✅ | ❌ |
| Simpler Model | ❌ | ❌ | ✅ |

---

# Real-World Example

Suppose we want to predict student exam scores using:

- Hours Studied
- Attendance
- Sleep Hours
- Shoe Size
- Favorite Color

Linear Regression may use all features.

Ridge Regression will reduce the impact of less important features but keep them in the model.

Lasso Regression will reduce the impact of less important features and may completely remove features such as Shoe Size and Favorite Color if they do not help the prediction.

As a result, Lasso often produces a cleaner and simpler model.

---

# Summary

Linear Regression focuses only on reducing prediction error.

Ridge Regression focuses on reducing prediction error while keeping coefficients small.

$$
Ridge = Error + L2\ Penalty
$$

Lasso Regression focuses on reducing prediction error while shrinking coefficients and removing unnecessary features.

$$
Lasso = Error + L1\ Penalty
$$

Ridge Regression keeps all features but reduces their impact.

Lasso Regression can automatically remove unimportant features by making their coefficients equal to zero.

Both techniques are widely used to improve model performance and reduce overfitting in machine learning.
