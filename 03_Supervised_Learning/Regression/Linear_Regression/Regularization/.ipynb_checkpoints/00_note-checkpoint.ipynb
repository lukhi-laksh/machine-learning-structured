# Lasso Regression and Ridge Regression (Detailed Theory)

## Introduction

Lasso Regression and Ridge Regression are advanced versions of Linear Regression that are used to reduce overfitting and improve model performance.

Both methods belong to a family of techniques called **Regularization**.

The main purpose of regularization is to prevent a machine learning model from becoming too complex and learning unnecessary patterns from training data.

A good machine learning model should not only perform well on training data but should also perform well on unseen data.

---

# The Real Problem with Linear Regression

Linear Regression tries to find the relationship between input features and the target variable.

The Linear Regression equation is:

$$
y = b_0 + b_1x_1 + b_2x_2 + \cdots + b_nx_n
$$

Where:

- $y$ = Predicted Value
- $b_0$ = Intercept
- $b_1,b_2,\dots,b_n$ = Coefficients
- $x_1,x_2,\dots,x_n$ = Features

The goal of Linear Regression is to find coefficient values that minimize prediction error.

Its cost function is:

$$
Cost = \sum (y - \hat{y})^2
$$

Where:

- $y$ = Actual Value
- $\hat{y}$ = Predicted Value

Linear Regression only focuses on reducing error.

It does not care whether a feature is useful or not.

It simply adjusts coefficients until prediction error becomes as small as possible.

---

# Understanding Overfitting

Suppose we want to predict Exam Scores.

Features:

- Hours Studied
- Attendance
- Sleep Hours
- Shoe Size
- Favorite Color

A Linear Regression model may produce:

$$
ExamScore =
5
+
8(HoursStudied)
+
3(Attendance)
+
2(SleepHours)
+
1.5(ShoeSize)
+
0.5(FavoriteColor)
$$

Notice that Shoe Size and Favorite Color have almost no logical relationship with Exam Score.

However, Linear Regression may still use them if they help reduce training error.

This is called:

## Overfitting

Overfitting occurs when a model starts learning noise instead of learning the true relationship.

As a result:

- Training Accuracy becomes very high.
- Testing Accuracy becomes low.
- Model performs poorly on new data.

---

# Solution: Regularization

Scientists introduced a penalty system.

Instead of minimizing only error, the model now minimizes:

$$
Cost = Error + Penalty
$$

The model now has two objectives:

1. Reduce Prediction Error
2. Keep Coefficients Small

This process is called:

## Regularization

Regularization prevents coefficients from becoming excessively large.

This reduces model complexity and helps prevent overfitting.

---

# What is Ridge Regression?

Ridge Regression is a regularized version of Linear Regression that adds a penalty based on the square of coefficients.

Its cost function is:

$$
Cost =
RSS
+
\lambda \sum b_i^2
$$

Expanded form:

$$
Cost =
RSS
+
\lambda
(b_1^2+b_2^2+b_3^2+\cdots+b_n^2)
$$

Where:

- $RSS$ = Residual Sum of Squares
- $\lambda$ = Regularization Parameter
- $b_i$ = Model Coefficients

This penalty is called:

## L2 Regularization

---

# Why Does Ridge Use Squared Coefficients?

Suppose we have two coefficients:

$$
2^2 = 4
$$

$$
10^2 = 100
$$

The larger coefficient receives a much larger penalty.

Because of this:

- Large coefficients are heavily punished.
- Small coefficients receive little punishment.

This forces the model to keep coefficient values smaller.

---

# Backend Working of Ridge Regression

Suppose Linear Regression produces:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 15 |
| Attendance | 12 |
| Sleep Hours | 8 |

Penalty becomes:

$$
15^2 + 12^2 + 8^2
$$

$$
=
225 + 144 + 64
$$

$$
=
433
$$

This penalty is added to the cost function.

The optimizer now tries to reduce both:

- Prediction Error
- Penalty

After several iterations:

$$
15 \rightarrow 12 \rightarrow 10
$$

$$
12 \rightarrow 10 \rightarrow 8
$$

$$
8 \rightarrow 6 \rightarrow 5
$$

Final Model:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 10 |
| Attendance | 8 |
| Sleep Hours | 5 |

The coefficients become smaller.

The model becomes less complex.

Overfitting decreases.

---

# Ridge Regression Training Algorithm

### Step 1

Initialize coefficients randomly.

$$
b_0,b_1,b_2,\cdots,b_n
$$

### Step 2

Calculate predictions.

$$
\hat{y}
=
b_0+b_1x_1+b_2x_2+\cdots+b_nx_n
$$

### Step 3

Calculate Error.

$$
RSS
=
\sum(y-\hat{y})^2
$$

### Step 4

Calculate Ridge Cost.

$$
Cost
=
RSS
+
\lambda \sum b_i^2
$$

### Step 5

Calculate Gradients.

### Step 6

Update Coefficients.

$$
b_{new}
=
b_{old}
-
\alpha
\frac{\partial Cost}{\partial b}
$$

Where:

$$
\alpha
=
Learning\ Rate
$$

### Step 7

Repeat until convergence.

---

# Why Ridge Does Not Remove Features

Ridge continuously shrinks coefficients.

Example:

$$
5
\rightarrow
4
\rightarrow
3
\rightarrow
2
\rightarrow
1
\rightarrow
0.5
\rightarrow
0.1
\rightarrow
0.01
$$

The coefficient becomes extremely small.

However, it rarely becomes exactly:

$$
0
$$

Therefore:

- Ridge keeps all features.
- Ridge only reduces their influence.

---

# What is Lasso Regression?

Lasso Regression is another regularized version of Linear Regression.

Lasso stands for:

**Least Absolute Shrinkage and Selection Operator**

Its cost function is:

$$
Cost
=
RSS
+
\lambda \sum |b_i|
$$

Expanded form:

$$
Cost
=
RSS
+
\lambda
(
|b_1|
+
|b_2|
+
|b_3|
+
\cdots
+
|b_n|
)
$$

This penalty is called:

## L1 Regularization

---

# Why Lasso Can Remove Features

Suppose Linear Regression produces:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 15 |
| Attendance | 12 |
| Sleep Hours | 8 |
| Shoe Size | 1 |
| Favorite Color | 0.5 |

Notice:

- Hours Studied is important.
- Attendance is important.
- Sleep Hours is important.
- Shoe Size contributes very little.
- Favorite Color contributes very little.

Lasso decides:

> Keeping these tiny coefficients increases penalty but barely improves prediction.

Therefore it removes them.

Final result:

| Feature | Coefficient |
|----------|----------|
| Hours Studied | 10 |
| Attendance | 8 |
| Sleep Hours | 5 |
| Shoe Size | 0 |
| Favorite Color | 0 |

The removed features no longer affect predictions.

---

# How Lasso Decides a Feature is Useless

Many beginners think Lasso somehow "knows" which feature is important.

It does not.

It only minimizes cost.

Suppose:

### Keeping Feature

$$
Error = 100
$$

$$
Penalty = 20
$$

$$
TotalCost = 120
$$

### Removing Feature

$$
Error = 102
$$

$$
Penalty = 15
$$

$$
TotalCost = 117
$$

Since:

$$
117 < 120
$$

The optimizer prefers removing that feature.

This is how automatic feature selection occurs.

No artificial intelligence is involved.

Only mathematical optimization.

---

# Lasso Regression Training Algorithm

### Step 1

Initialize coefficients.

$$
b_0,b_1,b_2,\cdots,b_n
$$

### Step 2

Calculate predictions.

$$
\hat{y}
=
b_0+b_1x_1+b_2x_2+\cdots+b_nx_n
$$

### Step 3

Calculate Error.

$$
RSS
=
\sum(y-\hat{y})^2
$$

### Step 4

Calculate Lasso Cost.

$$
Cost
=
RSS
+
\lambda \sum |b_i|
$$

### Step 5

Calculate Gradients.

### Step 6

Update Coefficients.

$$
b_{new}
=
b_{old}
-
\alpha
\frac{\partial Cost}{\partial b}
$$

### Step 7

Small coefficients move toward:

$$
0
$$

### Step 8

Features with coefficients equal to zero are removed.

### Step 9

Repeat until convergence.

---

# Understanding Lambda ($\lambda$)

Lambda controls the strength of regularization.

---

## Small Lambda

$$
\lambda \approx 0
$$

Effects:

- Very little regularization.
- Model behaves like Linear Regression.
- Risk of overfitting remains.

---

## Large Lambda

$$
\lambda \gg 0
$$

Effects:

- Strong regularization.
- Coefficients become much smaller.
- Model becomes simpler.
- Too much regularization can cause underfitting.

---

# Why Feature Scaling is Important

Suppose:

| Feature | Range |
|----------|----------|
| Age | 1 - 100 |
| Salary | 1 - 1,000,000 |

Without scaling:

- Salary dominates the model.
- Penalty becomes unfair.

Therefore before using:

- Ridge Regression
- Lasso Regression

We usually apply:

$$
StandardScaler
$$

or

$$
MinMaxScaler
$$

---

# Geometric Understanding

Ridge Regression creates a circular constraint.

Because circles are smooth, coefficients usually shrink toward zero but rarely become exactly zero.

Lasso Regression creates a diamond-shaped constraint.

The corners of the diamond lie on the axes.

Optimization frequently reaches those corners.

Therefore coefficients become:

$$
0
$$

exactly.

This is the mathematical reason Lasso performs feature selection.

---

# Elastic Net Regression

Elastic Net combines Ridge and Lasso.

Cost Function:

$$
Cost
=
RSS
+
\lambda_1\sum |b_i|
+
\lambda_2\sum b_i^2
$$

Elastic Net provides:

- Feature Selection from Lasso
- Stability from Ridge

---

# Difference Between Linear, Ridge and Lasso Regression

| Property | Linear | Ridge | Lasso |
|----------|----------|----------|----------|
| Reduces Overfitting | ❌ | ✅ | ✅ |
| Shrinks Coefficients | ❌ | ✅ | ✅ |
| Removes Features | ❌ | ❌ | ✅ |
| Feature Selection | ❌ | ❌ | ✅ |
| Handles Multicollinearity | ❌ | ✅ | ✅ |
| Keeps All Features | ✅ | ✅ | ❌ |
| Simpler Model | ❌ | ❌ | ✅ |

---

# Easy Memory Trick

Imagine coefficients are employees in a company.

### Linear Regression

Keeps everyone.

$$
Cost = RSS
$$

---

### Ridge Regression

Cuts everyone's salary but keeps everyone employed.

$$
Cost
=
RSS
+
\lambda\sum b_i^2
$$

---

### Lasso Regression

Fires employees who contribute very little.

$$
Cost
=
RSS
+
\lambda\sum |b_i|
$$

---

### Elastic Net

Cuts salaries and also fires unnecessary employees.

$$
Cost
=
RSS
+
\lambda_1\sum |b_i|
+
\lambda_2\sum b_i^2
$$

---

# Final Summary

Linear Regression only minimizes prediction error.

$$
Cost = RSS
$$

Ridge Regression minimizes prediction error while shrinking coefficients.

$$
Cost
=
RSS
+
\lambda\sum b_i^2
$$

Lasso Regression minimizes prediction error while shrinking coefficients and removing unimportant features.

$$
Cost
=
RSS
+
\lambda\sum |b_i|
$$

Ridge keeps all features but reduces their impact.

Lasso automatically removes features that contribute very little to prediction.

Both techniques are powerful tools for reducing overfitting and improving machine learning model performance.
