# K-Nearest Neighbours (KNN)

### KNN is a non-parametric, instance-based learning algorithm used for:
##### Classification (labels like Yes/No)

##### Regression (continuous values like price, marks)

##### KNN does NOT learn a model.
##### It stores the entire dataset and works only at prediction time.

KNN predicts the output for a new data point by finding the K closest data points in the training data and using their values to make a decision.

##### **Why KNN is Non-Linear?**

**Because:**
- It does not create a straight-line equation
- Predictions depend on local neighbors
- Decision boundaries are irregular & curved
- So → Non-Linear Algorithm

## Backend Working of KNN

## 🧠 Step-by-Step Backend Process of KNN

### Step 1: Choose K
$$
K = \text{number of nearest neighbors}
$$

$$
K = 3 \Rightarrow \text{use 3 closest points}
$$

$$
K = 5 \Rightarrow \text{use 5 closest points}
$$

---

### Step 2: Calculate Distance
$$
\text{Distance is calculated between the new data point and all training data points}
$$

$$
\text{Most commonly used distance: Euclidean Distance}
$$

---

### Euclidean Distance Formula

#### For one feature (1D):
$$
d = \sqrt{(x_2 - x_1)^2}
$$

#### For multiple features (nD):
$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + \cdots}
$$

---

### Step 3: Sort Distances
$$
d_1 \le d_2 \le d_3 \le \cdots \le d_n
$$

---

### Step 4: Select K Nearest Points
$$
\{d_1, d_2, \ldots, d_K\}
$$

---

### Step 5: Make Prediction

#### Classification (Majority Vote):
$$
\hat{y} = \text{mode}(y_1, y_2, \ldots, y_K)
$$

#### Regression (Average):
$$
\hat{y} = \frac{1}{K} \sum_{i=1}^{K} y_i
$$

---

## 🔹 KNN Regression Example

### Dataset: Study Hours vs Marks
$$
\begin{array}{|c|c|}
\hline
\text{Study Hours} & \text{Marks} \\
\hline
1 & 35 \\
2 & 40 \\
4 & 55 \\
6 & 70 \\
8 & 85 \\
\hline
\end{array}
$$

---

### Predict Marks for Study Hours = 5
$$
x_{\text{new}} = 5, \quad K = 3
$$

---

### Distance Calculation
$$
d(5,1) = |5 - 1| = 4
$$

$$
d(5,2) = |5 - 2| = 3
$$

$$
d(5,4) = |5 - 4| = 1
$$

$$
d(5,6) = |5 - 6| = 1
$$

$$
d(5,8) = |5 - 8| = 3
$$

---

### Sorted Distances
$$
\begin{array}{|c|c|c|}
\hline
\text{Study Hours} & \text{Marks} & \text{Distance} \\
\hline
4 & 55 & 1 \\
6 & 70 & 1 \\
2 & 40 & 3 \\
\hline
\end{array}
$$

---

### Final Prediction
$$
\hat{y} = \frac{55 + 70 + 40}{3}
$$

$$
\hat{y} = 55
$$

---

### Result
$$
\boxed{\text{Predicted Marks} = 55}
$$


```python
Marks
 90 |                     ● (8,85)
 80 |
 70 |             ● (6,70)
 60 |         X  ← Predict (5,?)
 50 |         ● (4,55)
 40 |     ● (2,40)
 30 | ● (1,35)
     ----------------------------
       1   2   3   4   5   6   7   8
              Study Hours
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
```

```python
X = np.array([[1], [2], [4], [6], [8]])
y = np.array([35, 40, 55, 70, 85])
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X, y)
```

```python
pred = knn.predict([[5]])
print("Predicted Marks:", pred[0])
```

```python
X_test = np.arange(1, 9, 0.1).reshape(-1, 1)

plt.scatter(X, y)
plt.plot(X_test, knn.predict(X_test))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("KNN Regression (Non-Linear)")
plt.show()
```
