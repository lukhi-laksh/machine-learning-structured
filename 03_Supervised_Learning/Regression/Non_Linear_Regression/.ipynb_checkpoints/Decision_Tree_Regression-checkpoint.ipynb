# Decision Tree 

### How Decision Tree Works

- It looks at all features
- Finds the best question to split data
  
**Example:**
- Is study_hours ≤ 5?
- Is experience ≤ 3 years?
- Splits data into branches

**Keeps splitting until:**
- Data becomes very similar
- Or max depth is reached
- Final node gives average output value

### **Why Decision Tree is Non-Linear?**

###### **Because:**

- It does not draw a straight line

- It makes many splits

- The final shape looks like steps, not a line

- So it can learn complex patterns easily.

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.tree import DecisionTreeRegressor
```

```python
X = np.array([[1], [2], [4], [6], [8]])   # Study Hours (input)
y = np.array([35, 40, 55, 70, 85])        # Marks (output)
```

```python
model = DecisionTreeRegressor(max_depth=3)
model.fit(X, y)
```

```python
study_hours = [[3], [5], [7]]
predictions = model.predict(study_hours)

print("Predicted Marks:", predictions)
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(-1, 1)

plt.scatter(X, y)
plt.plot(X_grid, model.predict(X_grid))
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Decision Tree Regression (Non-Linear)")
plt.show()
```

**Output Explanation:**

- Model splits data into steps

- Prediction changes block-wise, not smoothly

- This shows non-linear regression
