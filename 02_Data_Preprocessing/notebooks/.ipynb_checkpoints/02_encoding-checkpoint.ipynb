# Encoding

- Machine learning models work only with numbers
- Categorical features cannot be used directly
- Encoding helps preserve information in numeric form

## 1. Dummy Encoding

```python
# when we have only limited catogrical data like gender, (yes, no), citizen then  that kind of data on we can use Ordinal Encoding
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
```

### 1. using Pandas "get_dummis() method"

```python
en_data = df[["Gender", "Married"]]
en_data
```

```python
dummies = pd.get_dummies(en_data)
dummies
```

```python
# Here we can show that show us data in boolean 
dummies.info()
```

```python
# but we want to our data in Numerical form so we use scikit learn library so .... here is method...
```

## 2. One-Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
data = ohe.fit_transform(en_data).toarray()
```

```python
print(data)
```

```python
pd.DataFrame(data, columns=dummies.columns)
```

```python
# OneHotEncoder return float value.
# what if we have 5 different data in our column then that OneHotEncoder generate 5 new columns
# that OneHotEncoder convert our dataset larger to larger
# what if we have one column like gender where one two different data in one column then if we get only one columns in data then we
#        know another value so "drop = 'first'" remove first one means now it is remove first column.
```

```python
en_data.head()
```

```python
ohe = OneHotEncoder(drop='first')
data_drop = ohe.fit_transform(en_data).toarray()
data_drop
```

```python
# We can show here both of columns first column is delete
# Now we can get same information in two column if there is 0 then we just assume that have female or not merried

pd.DataFrame(data_drop, columns=['Gender_Male', 'Married_Yes'])
```

## 3. Label Encoding

```python
# when we need to give number in assending order in series then we are use labelEncoding
df = pd.DataFrame({"name":['Laksh', 'Cow', 'Laksh', 'Dog', 'Black']})
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Asc'] = le.fit_transform(df['name'])
```

```python
# now this new column Asc contains assending numbers / start from 0
# so using that label we can write all unique things identify with numbers
df
```

```python
label_test = pd.read_csv('../../Dataset/loan.csv')
label_test.head(3)
```

```python
label_test['Property_Area'].unique()
```

```python
temp = le.fit(label_test['Property_Area'])
temp
```

```python
label_test['Property_Area'] = le.transform(label_test['Property_Area'])
```

```python
# Now we can show all of them gives the numbers in assending order 
label_test['Property_Area']
```

```python
# array(['Urban', 'Rural', 'Semiurban'], dtype=object) that convers in number in asc oders.
label_test['Property_Area'].unique()
```

## 4. Ordinal Encoding

```python
# there are two way to do that:
#    1. Cyclic Line
#    2. Map function
```

### i. Cyclic Line

```python
# first we just make simple dataset for better understanding then we just apply on real or large dataset

df = pd.DataFrame({"Size":['s', 'm','l', 'xl', 's', 'm', 'l', 's', 's', 'l', 'xl', 'm']})
df.head(3)
```

```python
ord_data = [['s', 'm', 'l', 'xl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python
df["size_oe"] = oe.transform(df[['Size']])
```

```python
df.head()
```

### ii. Using Map 

```python
ord_data1 = {"s": 0, "m":1, "l":2, "xl":6}
df['size_map'] = df['Size'].map(ord_data1)
df
```

```python
# now we do Ordinal encoding with large dataset
label_test = pd.read_csv('../../Dataset/loan.csv')
label_test.head(3)
```

### iii. using order list

```python
en_data_ord = [['Urban', 'Rural', 'Semiurban']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oen = OrdinalEncoder(categories=en_data_ord)
```

```python
label_test['Property_Area'] = oen.fit_transform(label_test[['Property_Area']])
```
