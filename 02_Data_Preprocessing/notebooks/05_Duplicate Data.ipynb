# Duplicate data

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data = {"name":['a', 'b', 'c', 'd', 'a', 'c'], "eng":[8, 7, 6, 2, 8, 6], "hindi":[2, 3, 5, 9, 2, 5]}

df = pd.DataFrame(data)
df
```

```python
df['duplicated'] = df.duplicated()
df
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
df = pd.read_csv('../../Dataset/loan.csv')
df.head(3)
```

```python
# here we can show 'Dependents' Column have numeric value it's be int or floting format so we need to convert it
df.info()
```

```python
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
```

```python
df['Dependents'].unique()
```

```python
df['Dependents'].value_counts()
```

```python
df['Dependents'].replace("3+","3", inplace=True)
```

```python
df['Dependents'].value_counts()
```

```python
# it is still show Object Datatype means we should need convert it to int
df['Dependents'].info()
```

```python
df['Dependents'] = df['Dependents'].astype("int64")
```
