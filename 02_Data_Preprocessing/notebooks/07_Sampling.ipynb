# Imbalanced Data

```python
# Imbalanced data in data science means the number of samples in each class is not equal.
# One class has many more examples, and another class has very few.

# The model becomes biased if there is only few catagory any one
# video ex. Airports on indian People
```

### Example

```python
# Fraud detection → 99% normal, 1% fraud
# Medical diagnosis → 95% healthy, 5% sick
# Spam detection → 90% normal emails, 10% spam
```

## Techniques to handle inbalanced Data
### 1. **Random Under Sampling:** We will reduce the majority of the class so that it will have same no of as the minority
### 2. **Random Over Sampling:** We will increase the size of manority is inactive class to the size of majority class ie active.

## **1. Random Under Sampling**

```python
# Random Under Sampling (RUS)

# Suppose you have imbalanced data:
# Example:
#   900 samples of class 0
#   100 samples of class 1
#
# The dataset is unbalanced because one class (0) is much larger.

# What Random Under Sampling does:
# - It removes (deletes) some data from the majority class (class 0)
# - It reduces the big class down to the size of the smaller class
#
# Example:
#   From 900 samples of class 0 → randomly keep only 100 samples
#   Class 1 already has 100 samples → keep all
#
# After RUS:
#   100 samples of class 0
#   100 samples of class 1
#   Balanced dataset!

# Why we use it:
# - To make machine learning models learn equally from both classes
# - To avoid bias toward the majority class

# Downside:
# - We lose some useful data because many samples of the majority class are removed.
```

## **2. Random Over Sampling**

```python
# Random Over Sampling (ROS)

# Suppose your dataset is imbalanced:
# Example:
#   900 samples of class 0 (majority)
#   100 samples of class 1 (minority)

# What Random Over Sampling does:
# - It increases the number of samples in the minority class (class 1)
# - It does this by randomly repeating (copying) some of the minority samples
#
# Example:
#   Class 1 has 100 samples → randomly duplicate them until it becomes 900
#   Class 0 already has 900 → keep as it is

# After ROS:
#   900 samples of class 0
#   900 samples of class 1
#   Balanced dataset!

# Why we use it:
# - To give equal importance to both classes during training
# - It helps the model learn patterns from the minority class

# Downside:
# - Because it duplicates existing data, it can cause overfitting
#   (model learns the same samples again and again)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
```

```python
ro_x, ro_y = ro.fit_resample(x, y)
```

```python
print(ro_y.value_counts())
```

```python
# In Over Sampling Repete our small amount of data Repete and repeted it to Almost same value
```

```python
x_train, x_test, y_train, y_test = train_test_split(ro_x, ro_y, test_size=0.2, random_state=42)
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[46, 41000]])
```
