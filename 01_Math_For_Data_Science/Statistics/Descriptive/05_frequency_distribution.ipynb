# Frequency Distribution

Frequency Distribution is a statistical technique used to organize data by showing how often each value or group of values occurs in a dataset.

## Types of Frequency Distribution

### 1. Ungrouped Frequency Distribution
- Used for small datasets
- Counts frequency of each individual value

**Example:**

```python
Data = [2, 4, 4, 6, 2, 4]
pd.Series(Data).value_counts()
```

### 2. Grouped Frequency Distribution
- Used for large datasets
- Data is divided into class intervals

**Example:**

```python
Data = [10, 15, 18, 22, 25, 28, 35, 40, 45, 48]
df = pd.DataFrame({'Data': Data})

bins = [10, 20, 30, 40, 50]
labels = ['10–20', '21–30', '31–40', '41–50']

grouped_freq = pd.cut(df['Data'], bins=bins, labels=labels).value_counts().sort_index()
grouped_freq
```

### Explanation
##### pd.cut() divides numerical data into class intervals
##### value_counts() counts values in each interval
##### sort_index() keeps intervals in correct order
